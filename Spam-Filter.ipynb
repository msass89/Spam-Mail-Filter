```python
import zipfile
```

```python
with zipfile.ZipFile('data/Spam-Emails.zip', mode = 'r') as z:
    z.extractall('data/Spam-Emails')
```

```python
def get_email_content(path):
    with open(path, mode = "r", encoding = "iso-8859-1") as file:
        email = file.read()
        email_parts = email.split("\n\n", maxsplit = 1)
        if len(email_parts) != 2:
            return None
        else:
            content = email_parts[1]
            return content
```

```python
import os

def read_email(folder):
    mail_content = []
    for file in os.listdir(folder):
        complete_path = folder + '/' + file
        if os.path.isdir(complete_path):
            continue
        content = get_email_content(complete_path)
        if content != None:
            mail_content.append(content)
    return mail_content

spam_mails = read_email('data/Spam-Emails/spam')
ham_mails = read_email('data/Spam-Emails/easy_ham')
```

```python
print(len(spam_mails))
print(len(ham_mails))
```

```python
def email_is_spam(mail):
        if mail.lower().count("click") == 0:
            return False
        elif mail.lower().count("offer") != 0:
            return True
        elif mail.lower().count("million") != 0:
            return True
        else:
            return True
```

```python
#Spam-Mails, 1 means that we correctly recognized the e-mail as spam
spam_detected = [int(email_is_spam(mail)) for mail in spam_mails]

#Non-Spam-E-Mails, 1 means that we mistakenly recognized it as an Spam-E-Mail
ham_spam_detected = [int(email_is_spam(mail)) for mail in ham_mails]

#The closer the result is near to 1, the more accurate we recognized all Spam-Mails 
print(sum(spam_detected) / len(spam_mails))

#The closer the result is near to 0, the more accurate we recognized Non-Spam-Mails as Non-Spam
print(sum(ham_spam_detected) / len(ham_mails))
```
